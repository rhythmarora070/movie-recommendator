```python
import numpy as np
import pandas as pd
```

```python
credits = pd.read_csv('tmdb_5000_credits[1].csv')
movies = pd.read_csv('tmdb_5000_movies[1].csv')
```

```python
movies.head(1)
```

```python
movies.info()
```

```python
movies = movies.merge(credits, on='title')
```

```python
# columns required
# genres
# id
# keywords
# title
# overview
# cast
# crew - because of director
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast # converts string of list to list
def convert(n):
  l = []
  for i in ast.literal_eval(n):
    l.append(i['name'])
  return l
```

```python
movies.genres = movies.genres.apply(convert)
```

```python
movies.keywords
```

```python
movies.keywords = movies.keywords.apply(convert)
```

```python
movies.iloc[0].cast
```

```python
def convert_cast(n):
  l = []
  count = 0
  for i in ast.literal_eval(n):
    if count != 3:
      l.append(i['name'])
      count += 1
  return l
```

```python
movies.cast = movies.cast.apply(convert_cast)
```

```python
def fetch_director(n):
  l = []
  for i in ast.literal_eval(n):
    if i['job'] == 'Director':
      l.append(i['name'])
  return l
```

```python
movies.crew = movies.crew.apply(fetch_director)
```

```python
movies.iloc[0].overview
```

```python
movies.overview = movies.overview.apply(lambda x:x.split())
```

```python
movies.genres = movies.genres.apply(lambda x:[i.replace(" ","")for i in x])
movies.cast = movies.cast.apply(lambda x:[i.replace(" ","")for i in x])
movies.crew = movies.crew.apply(lambda x:[i.replace(" ","")for i in x])
movies.keywords = movies.keywords.apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
movies['tag'] = movies.overview + movies.genres + movies.keywords + movies.cast + movies.crew
```

```python
movies.head()
```

```python
new_df = movies[['movie_id', 'title', 'tag']]
```

```python
new_df.tag = new_df.tag.apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df.tag = new_df.tag.apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
!pip install nltk
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df.tag = new_df.tag.apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')

vector = cv.fit_transform(new_df['tag']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse = True, key = lambda x:x[1])[2:7]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Spectre')
```
